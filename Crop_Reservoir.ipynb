## Resize all geotiff

```python
def resizeRasterObject(oldFilename, newFilename, shapeObject):
    rObject = rasterio.open(oldFilename, 'r')
    rData, originalTransform = getMaskFromShape(shapeObject, rObject)
    crs = rObject.crs
    dataType = rObject.dtypes[0]
    rObject.close()

    rObject = rasterio.open(newFilename, 'w', driver='GTiff',
                  height=rData.shape[0],width=rData.shape[1],count=1,
                  dtype=dataType,crs=crs,transform=originalTransform)
    rObject.write(rData, 1)
    rObject.close()
```

```python
root = 'Landsat8/'

for _, folders, __ in os.walk(root):
    break

c = 0
for dir in folders:
    c += 1
    print("[{0}/{1}] Processing {2}...".format(c, len(folders), dir))
    fnBase = os.path.join(root, dir, dir) + '_###' + '.TIF'
    for b in range(11):
        fn = fnBase.replace('###', 'B' + str(b+1))
        try:
            resizeRasterObject(fn, fn, ho_tri_an)
        except Exception as e:
            print(" --> Resizing {0} ... ".format(fn), end='')
            print(" -> Error: {0}".format(e))
    for newPostPrefix in ['BQA', 'NDWI', 'NDWI2', 'NDWI3']:
        fn = fnBase.replace('###', newPostPrefix)
        try:
            resizeRasterObject(fn, fn, ho_tri_an)
        except Exception as e:
            print(" --> Resizing {0} ... ".format(fn), end='')
            print(" -> Error: {0}".format(e))
```
