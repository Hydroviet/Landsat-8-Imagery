```python
import json
import os
```

```python
for root, dirs, filenames in os.walk('L7_Data/'):
    break
    
area = []
date = []

for folder in dirs:
    dPath = os.path.join('L7_Data', folder, 'info.json')
    try:
        jsonData = json.load(open(dPath, 'r'))
        area.append(float(jsonData['area']))
        date.append(jsonData['date'])
    except Exception as e:
        print(e)
```

```python
from matplotlib import pyplot
pyplot.figure(num=None, figsize=(15, 7), dpi=80, facecolor='w', edgecolor='k')
pyplot.plot(date,area)
pyplot.show()
```

```python
from plotly.graph_objs import *
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import numpy as np

init_notebook_mode(connected=True)

_date = []
_value = []

for i in range(len(date)):
    y, m, d = date[i].split('_')
    if (y == '2010'):
        _value.append(area[i])
        _date.append('_'.join([y, m, d]))
        
iplot([Scatter(x=_date,y=_value)])
```
